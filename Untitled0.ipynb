```python
# Installation automatique
!pip install matplotlib pandas numpy pillow

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
import numpy as np
from IPython.display import HTML

# Données
data = {
    'Année': list(range(2015, 2026)),
    'Chine': [4500, 4650, 4808, 4968, 5137, 5220, 5400, 5586, 5776, 5973, 6175],
    'États-Unis': [2000, 2040, 2081, 2122, 2165, 2180, 2240, 2305, 2374, 2445, 2518],
    'Inde': [1800, 1840, 1881, 1923, 1965, 1980, 2040, 2104, 2173, 2243, 2316],
    'Russie': [900, 918, 936, 955, 974, 980, 1010, 1040, 1071, 1104, 1138],
    'Iran': [500, 510, 520, 530, 541, 545, 560, 575, 590, 605, 620],
    'Japon': [700, 707, 714, 721, 728, 730, 750, 772, 795, 819, 843],
    'Allemagne': [500, 510, 520, 531, 541, 545, 560, 575, 590, 605, 620],
    'Corée du Sud': [400, 408, 416, 425, 433, 435, 445, 456, 468, 480, 492],
    'Royaume-Uni': [300, 306, 312, 319, 325, 327, 335, 345, 355, 366, 377],
    'Brésil': [400, 408, 416, 425, 433, 435, 445, 456, 468, 480, 492]
}

df = pd.DataFrame(data)

# Configuration
fig, ax = plt.subplots(figsize=(12, 7))
colors = plt.cm.Set3(np.linspace(0, 1, 10))

def animate(frame):
    ax.clear()
    year = df['Année'].iloc[frame]
    current_data = df[df['Année'] == year].iloc[0, 1:]

    # Tri pour l'affichage
    sorted_idx = current_data.sort_values(ascending=False).index
    sorted_values = current_data[sorted_idx]

    # Barres
    bars = ax.bar(range(len(sorted_values)), sorted_values.values,
                  color=[colors[i] for i in range(len(sorted_values))])

    # Labels
    ax.set_xticks(range(len(sorted_values)))
    ax.set_xticklabels([str(idx) for idx in sorted_idx], rotation=45, ha='right')

    ax.set_title(f'Nombre d\'Ingénieurs par Pays - {year}', fontsize=16, fontweight='bold')
    ax.set_ylabel('Nombre (en milliers)')
    ax.set_ylim(0, max(sorted_values.values) * 1.1)

    # Valeurs sur les barres
    for i, bar in enumerate(bars):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 20,
                f'{int(height):,}', ha='center', va='bottom', fontsize=9)

    ax.grid(axis='y', alpha=0.3)

# Créer l'animation
anim = animation.FuncAnimation(fig, animate, frames=len(df), interval=1000, repeat=True)

# Afficher dans Colab
HTML(anim.to_jshtml())
# Nécessite ffmpeg (installation : https://ffmpeg.org/download.html)
anim.save('ingenieurs_2015_2025.mp4', writer='ffmpeg', fps=1,
          extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])

# Taille du fichier : ~1-3 MB
# Compatible : YouTube, Vimeo, présentations PowerPoint
```
